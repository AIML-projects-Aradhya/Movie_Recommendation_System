Importing Dependencies


```python
import numpy as np
import pandas as pd
import difflib #close search for names
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity #similary score for recommending
```

Data Collection and Pre-processing

```python
movie_data = pd.read_csv('/content/movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.isnull().sum()
```

```python
#selecting the relevant features
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
print(selected_features)
```

```python
#replacing the null variables with null string
for feature in selected_features:
  movie_data[feature] = movie_data[feature].fillna('')
```

```python
#combining features
combined_features = movie_data['genres']+' '+movie_data['keywords']+' '+movie_data['tagline']+' '+movie_data['cast']+' '+movie_data['director']
print(combined_features)
```

```python
#converting into feature vectors using tfidfvectorizer
vectorizer = TfidfVectorizer()

feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

Cosine Similary

```python
#similarity score
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting user input
movie_name = input('Enter your favourite movie name: ')
```

```python
#creating a list with all the movie names given in the dataset
titles = movie_data['title'].tolist()
print(titles)
```

```python
#finding the cross-match for user input
#using difflib
match = difflib.get_close_matches(movie_name, titles)
print(match)
closest_match = match[0]
print(closest_match)
```

```python
#index of movie
index = movie_data[movie_data.title == closest_match]['index'].values[0]
print(index)
```

```python
#getting list of similar values based on index
similarity_score = list(enumerate(similarity[index]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#refining in desc order
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
#print 10-20 similar movies based on above index
print('Movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index == index]['title'].values[0]
  if(i<10):
    print(i, '.', title_from_index)
    i+=1
```

Movie Recommendation System

```python
movie_name = input('Enter your favourite movie name: ')
titles = movie_data['title'].tolist()
match = difflib.get_close_matches(movie_name, titles)
closest_match = match[0]
index = movie_data[movie_data.title == closest_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index]))
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print('Movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index == index]['title'].values[0]
  if(i<10):
    print(i, '.', title_from_index)
    i+=1
```
